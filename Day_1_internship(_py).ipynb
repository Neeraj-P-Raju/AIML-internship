Write a function that accepts a number and returns True if it is prime, otherwise False

```python
def p(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
num = int(input("Enter a number: "))
print(p(num))
```

Count Odd Digits
Write a function that returns how many odd digits are present in a number

```python
n = input("Enter a number: ")
count=0
for dig in n:
    if int(dig)%2 != 0:
        count += 1
print("Odd digits =",count)
```
